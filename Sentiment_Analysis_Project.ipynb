```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(df.Review)
words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names_out())
words_df.head()
```

```python
m = words_df
n = df.Liked
```

```python
m
```

```python
n
```

```python
from sklearn.svm import LinearSVC
```

```python
%%time

svc = LinearSVC()
svc.fit(m, n)
```

```python
from sklearn.model_selection import train_test_split

m_train, m_test, n_train, n_test = train_test_split(m, n)
```

```python
%%time

print("Training SVC")
svc.fit(m_train, n_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
n_true = n_test
n_pred = svc.predict(m_test)
matrix = confusion_matrix(n_true, n_pred)

label_names = pd.Series(['negative', 'positive'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names)
```

```python
n_true = n_test
n_pred = svc.predict(m_test)
matrix = confusion_matrix(n_true, n_pred)

label_names = pd.Series(['negative', 'positive'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names).div(matrix.sum(axis=1), axis=0)
```
